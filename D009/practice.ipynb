```python
import requests
from bs4 import BeautifulSoup
```

```python
SITE_URL = 'https://news.naver.com/section/103'

response = requests.get("https://news.naver.com/section/103", timeout=5)

# 상태 코드 확인
print('상태 코드:', response.status_code)    # 200 = 정상
print('인코딩:   ', response.encoding)
print('HTML 길이:', len(response.text), '자')
print()
print('HTML 앞부분 미리보기:')
print(response.text[:200])
```

```python
soup = BeautifulSoup(response.text, "html.parser")
# selector = '#_SECTION_HEADLINE_LIST_bcr7h > li:nth-child(1) > div > div > div.sa_text > a > strong'
selector = 'ul[id^="_SECTION_HEADLINE_LIST"]'

headlines = soup.select_one(selector)
for i, li in enumerate(headlines.select("li"), 1):
    title = li.select_one("strong.sa_text_strong").get_text(strip=True)
    press = li.select_one("div.sa_text_press").get_text(strip=True)
    print(f"{i}. {title}, {press}")
```
